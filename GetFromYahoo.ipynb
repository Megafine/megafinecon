```python
import yfinance as yahooFinance


amazon = yahooFinance.Ticker("AMZN")

# Valid options are 1d, 5d, 1mo, 3mo, 6mo, 1y,
# 2y, 5y, 10y and ytd.
print(amazon.history(period="10y"))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.DataFrame(amazon.history(period="10y"))
df
```

```python
fig, ax = plt.subplots()
ax.plot(df.Close, label = "Close")
plt.xticks(rotation=45, ha='right')
plt.xlabel('Time', fontsize=12)
plt.ylabel('Price', fontsize=12)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df.Close-df.Open, label = "Close")
plt.xticks(rotation=45, ha='right')
plt.xlabel('Time', fontsize=12)
plt.ylabel('Price', fontsize=12)
ax.axhline(0, color='r')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(np.diff(df.Close), label = "Diff.Close")
plt.xticks(rotation=45, ha='right')
plt.xlabel('Time', fontsize=12)
plt.ylabel('Price', fontsize=12)
ax.axhline(0, color='r')
ax.legend()
plt.show()
```
